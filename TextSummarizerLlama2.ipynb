````python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer
import transformers
import torch
````

````python
model = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",#task
    model=model,
    tokenizer=tokenizer,
    torch_dtype = torch.bfloat16,
    trust_remote_code=True,
    device_map='auto',
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
````

````python
llm = HuggingFacePipeline(pipeline=pipeline,model_kwargs={'temperature':0})
````

````python
from langchain import PromptTemplate, LLMChain

template = """
              wrirte a concise summary of the following text delimited by triple backquotes.
              Return your response in an illustrative manner which sums up all the key points.
              And also try to present in bullet points.
              ```{text}```
              BULLET POINT SUMMARY:
            """
prompt = PromptTemplate(template=template,input_variables=["text"])
llm_chain = LLMChain(prompt=prompt,llm=llm)
text1 = """Lord Krishna, a central figure in Hindu mythology, is revered as the eighth avatar of the god Vishnu.
            Born in the Dwapara Yuga, Krishna's life is rich with divine exploits and teachings that continue to inspire millions worldwide.
            His childhood is adorned with tales of mischievousness, including stealing butter and playing pranks on the gopis (cowherd girls) in his village of Vrindavan.
            As he grew older, his wisdom and valor became evident, culminating in his role as a key strategist in the Mahabharata war, where he delivered the timeless scripture, the Bhagavad Gita, to his disciple Arjuna.
            Krishna's teachings in the Bhagavad Gita emphasize the principles of dharma (duty) and selflessness, providing profound insights into life's purpose and the path to spiritual fulfillment.
            He is often depicted with a flute, symbolizing the harmony he brings to the universe, and his divine love affair with Radha symbolizes the union of the individual soul (Atman) with the divine (Brahman).
            Lord Krishna's legacy transcends time, continuing to guide and inspire seekers on their spiritual journey.
        """
print(llm_chain.run(text1))
````
